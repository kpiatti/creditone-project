# Feature & Target Selection for Credit Limit Models

#### Re-order Columns
To make things easier, before idenifying my feature and target variables, I want to move the LIMIT_BAL var to the last column in my df. 

```python
#step 1: create a variabled to hold the LIMIT_BAL var
lim=df['LIMIT_BAL']

#step 2: drop the LIMIT_BAL column
df.drop(labels='LIMIT_BAL', axis=1, inplace=True)
```

```python
#step 3: insert 'lim' as last column of df
df.insert(loc=30, column='limit', value=lim, inplace=True)
```
